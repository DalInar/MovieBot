```python
PH_movies_data['Date'] = pd.to_datetime(PH_movies_data['Date'], format='%m/%d/%Y')
source = ColumnDataSource(PH_movies_data)

p = figure(x_axis_type='datetime',plot_width=800, plot_height=500)
p.line(x='Date', y='IMDB', 
         source=source, 
         line_width=2, line_dash='solid', color='green', legend='IMDB score')

# p.circle(x='Date', y='IMDB', 
#          source=source, 
#          size=5, color='green')

p.line(x='Date', y='META', 
         source=source, 
         line_width=2, line_dash='solid', color='blue', legend='Metascore')

# p.circle(x='Date', y='META', 
#          source=source, 
#          size=5, color='blue')

p.line(x='Date', y='PH_rating', 
         source=source, 
         line_width=2, line_dash='solid', color='red', legend='PH score')

# p.circle(x='Date', y='PH_rating', 
#          source=source, 
#          size=5, color='red')

p.title.text = 'Score Comparison'
p.xaxis.axis_label = 'Date'
p.yaxis.axis_label = 'Score'

hover = HoverTool()
hover.tooltips=[
    ('Movie', '@Movie'),
    ('Year', '@Year'),
    ('Rating', '@Rating')
]

p.add_tools(hover)
p.legend.click_policy='hide'

show(p)
```

```python
grouped = PH_movies_data.groupby('Rating')['IMDB', 'Movie'].count()
source = ColumnDataSource(grouped)
ratings = source.data['Rating'].tolist()
p = figure(x_range=ratings)

p.vbar(x='Rating', top='Movie', source=source, width=0.70)

p.title.text ='Ratings for Physics House Movies'
p.xaxis.axis_label = 'Rating'
p.yaxis.axis_label = 'Count'

hover = HoverTool()
hover.tooltips = [
    ("Totals", "@IMDB")]
hover.mode = 'vline'
p.add_tools(hover)

show(p)
```

```python
grouped = PH_movies_data.groupby('Year')['IMDB', 'Movie'].count()
source = ColumnDataSource(grouped)
ratings = source.data['Year'].tolist()
p = figure(x_range=ratings)

p.vbar(x='Year', top='Movie', source=source, width=0.70)

p.title.text ='Years Made for Physics House Movies'
p.xaxis.axis_label = 'Year'
p.yaxis.axis_label = 'Count'

hover = HoverTool()
hover.tooltips = [
    ("Count", "@IMDB"),("Year", "@Year")]
hover.mode = 'vline'
p.add_tools(hover)

show(p)
```

```python
grouped = PH_movies_data.groupby('DirectorGender')['IMDB', 'Movie'].count()
source = ColumnDataSource(grouped)
ratings = source.data['DirectorGender'].tolist()
ratings = [str(x) for x in ratings]
#p = figure(x_range=ratings)
p = figure()

p.vbar(x='DirectorGender', top='Movie', source=source, width=0.70)

p.title.text ='Director Gender for Physics House Movies'
p.xaxis.axis_label = 'Director Gender'
p.yaxis.axis_label = 'Count'

hover = HoverTool()
hover.tooltips = [
    ("Count", "@IMDB"),("Gender", "@DirectorGender")]
hover.mode = 'vline'
p.add_tools(hover)

show(p)
```
